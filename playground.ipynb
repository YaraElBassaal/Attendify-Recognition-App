```python
import os
import glob
```

```python
known_faces=[]
known_names=[]
known_faces_paths=[]

registered_faces_path='registered/'
for name in os.listdir(registered_faces_path):
    images_mask='%s%s/*.jpg'%(registered_faces_path ,name)
    images_paths=glob.glob(images_mask)
    known_faces_paths +=images_paths
    known_names+=[name for x in images_paths]
```

```python
for name,images_path in zip(known_names, known_faces_paths):
    print('%s : %s' %  (name, images_path ))
```

```python
import face_recognition
def get_encodings(img_path):
    image = face_recognition.load_image_file(img_path)
    encoding=face_recognition.face_encodings(image)
    return encoding[0]
```

```python
known_faces=[get_encodings(img_path) for img_path in known_faces_paths]
```

```python
import tkinter as tk
from tkinter import filedialog
import os
from tkinter import messagebox

image_path = ""

def upload_image():
    global image_path
    image_path = filedialog.askopenfilename(filetypes=[("Image Files", "*.png; *.jpg; *.jpeg")])
    # Do something with the selected image file
    import matplotlib.pyplot as plt

    # Assuming unknown_images is the variable containing the image path
    img = plt.imread(image_path)
    plt.figure()
    plt.imshow(img)

    encodings=face_recognition.face_encodings(img)
    found_faces = []
    for face_code in encodings:
        results=face_recognition.compare_faces(known_faces ,face_code ,tolerance=0.5)
        if any (results):
            found_faces.append(known_names[results.index(True)])
            plt.title('in the image : %s ' % str(found_faces))

    import xlsxwriter

    workbook = xlsxwriter.Workbook("Attendance.xlsx")
    worksheet = workbook.add_worksheet("attendance")

    worksheet.write(0, 0, 'id')

    for i in range(len(found_faces)):
        worksheet.write(i+1, 0,str(found_faces[i]))

    workbook.close()

def open_excel():
    file_path = "Attendance.xlsx"
    if os.path.exists(file_path):
        os.system(f'start excel "{file_path}"')
    else:
        messagebox.showerror("Error", "File 'Attendance.xlsx' not found.")

window = tk.Tk()
window.title("Image Uploader")

# Set the width and height of the window
window.geometry("400x400")

# Set the background color
window.configure(bg="#365486")

# Create the heading
heading_label = tk.Label(window, text="Attendance App", font=("Arial", 24), bg="#365486",foreground="#ffffff")
heading_label.pack(pady=20)

# Create a frame to hold the button
frame = tk.Frame(window, bg="#365486")
frame.pack()

# Create the upload button with some basic styling
upload_button = tk.Button(frame, text="Upload Image", command=upload_image, width=20, height=2, relief="solid", bd=2)
upload_button.pack(pady=20)

# Center the button horizontally
frame.grid_columnconfigure(0, weight=1)

open_button = tk.Button(window, text="Open Excel Sheet", command=open_excel, width=20, height=2, relief="solid", bd=2)
open_button.pack(padx=20, pady=20)

window.mainloop()





```
